2.8. работа с пропущенными значениями (если они есть): удаление строк с пропущенными значениями, заполнение пропущенных значений средним значением по колонке. Если пропущенных значений нет — намеренно их "генерируете", прибивая какие-то куски данных в np.nan

Сначала проверим наличие строк с пропущенными значениями

```python
students[students.isnull().any(axis=1)]
```

Запишем средние баллы по предметам в отдельную переменную для дальнейшей вставки в таблицу

```python
stud_mean = students.mean()
stud_mean
```

Добавим в таблицу 3 строки с нулевым значением и снова проверим наличие строк с пропущенными значениями

```python
students.loc[len(students.index)] = ['female', 'group C','some college','standard','completed',np.NaN,'76','82']
students.loc[len(students.index)] = ['female', 'group C','some college','standard','completed','54',np.NaN,'68']
students.loc[len(students.index)] = ['female', 'group C','some college','standard','completed','80','88',np.NaN]
```

```python
students[students.isnull().any(axis=1)]
```

Выведем на экран таблицу без пропущенных значений

```python
students.dropna()
```

Заполним пропущенные числовые значения средним арифметическим по столбцу и выведем новую таблицу на экран

```python
students["math score"] = students["math score"].fillna(value=stud_mean["math score"].mean())
students["reading score"] = students["reading score"].fillna(value=stud_mean["reading score"].mean())
students["writing score"] = students["writing score"].fillna(value=stud_mean["writing score"].mean())
```

```python
students
```

Далее удалим с таблицы строки со средними значениями и преобразуем столбцы с баллами с float в int

```python
students.drop(labels = [1000, 1001, 1002],axis = 0, inplace = True)
students["math score"] = students["math score"].astype(np.int64)
students["reading score"] = students["reading score"].astype(np.int64)
students["writing score"] = students["writing score"].astype(np.int64)
students
```
